```python
import re
def cleanResume(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)  
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requredTaxt  = tfidf.transform(df['Resume'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(requredTaxt, df['Category'], test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.svm import NuSVC
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
clf_1 = OneVsRestClassifier(KNeighborsClassifier())
clf_2 = KNeighborsClassifier()
clf_3 = MultinomialNB()
clf_4 = BernoulliNB()
clf_5 = GaussianNB()
clf_6 = LogisticRegression()
clf_7 = LinearSVC()
clf_8 = SVC()
clf_9 = SGDClassifier()
clf_1.fit(X_train , y_train)
clf_2.fit(X_train , y_train)
clf_3.fit(X_train , y_train)
clf_4.fit(X_train , y_train)
X_train_gauss = X_train.toarray()
clf_5.fit(X_train_gauss , y_train)
clf_6.fit(X_train , y_train)
clf_7.fit(X_train , y_train)
clf_8.fit(X_train , y_train)
clf_9.fit(X_train , y_train)
ypred_1 = clf_1.predict(X_test)
ypred_2 = clf_2.predict(X_test)
ypred_3  = clf_3.predict(X_test)
ypred_4  = clf_4.predict(X_test)
X_test_gauss = X_test.toarray()
ypred_5  = clf_5.predict(X_test_gauss)
ypred_6  = clf_6.predict(X_test)
ypred_7  = clf_7.predict(X_test)
ypred_8  = clf_8.predict(X_test)
ypred_9  = clf_9.predict(X_test)
print("ML Algorithm        " , "      Accuracy       " , "  Precision       " , "    Recall       " , "   F-Score       \n"    )
print("OnevsRestClassifier " , accuracy_score(y_test , ypred_1) , precision_score(y_test , ypred_1 , average = 'weighted'), recall_score(y_test , ypred_1 , average = 'weighted') , f1_score(y_test , ypred_1 , average = 'weighted'))
print("KNeighborsClassifier" , accuracy_score(y_test , ypred_2) , precision_score(y_test , ypred_2 , average = 'weighted'), recall_score(y_test , ypred_2 , average = 'weighted') , f1_score(y_test , ypred_2 , average = 'weighted'))
print("Multinomial NB      " , accuracy_score(y_test , ypred_3) , precision_score(y_test , ypred_3 , average = 'weighted'), recall_score(y_test , ypred_3 , average = 'weighted') , f1_score(y_test , ypred_3 , average = 'weighted'))
print("Bernoulli NB        " , accuracy_score(y_test , ypred_4) , precision_score(y_test , ypred_4 , average = 'weighted'), recall_score(y_test , ypred_4 , average = 'weighted') , f1_score(y_test , ypred_4 , average = 'weighted'))
print("Gaussian NB         " , accuracy_score(y_test , ypred_5) , precision_score(y_test , ypred_5 , average = 'weighted'), recall_score(y_test , ypred_5 , average = 'weighted') , f1_score(y_test , ypred_5 , average = 'weighted'))
print("Logistic Regression " , accuracy_score(y_test , ypred_6) , precision_score(y_test , ypred_6 , average = 'weighted'), recall_score(y_test , ypred_6 , average = 'weighted') , f1_score(y_test , ypred_6 , average = 'weighted'))
print("Linear SVC          " , accuracy_score(y_test , ypred_7) , precision_score(y_test , ypred_7 , average = 'weighted'), recall_score(y_test , ypred_7 , average = 'weighted') , f1_score(y_test , ypred_7 , average = 'weighted'))
print("SVC                 " , accuracy_score(y_test , ypred_8) , precision_score(y_test , ypred_8 , average = 'weighted'), recall_score(y_test , ypred_8 , average = 'weighted') , f1_score(y_test , ypred_8 , average = 'weighted'))
print("SGDClassifier       " , accuracy_score(y_test , ypred_9) , precision_score(y_test , ypred_9 , average = 'weighted'), recall_score(y_test , ypred_9 , average = 'weighted') , f1_score(y_test , ypred_9 , average = 'weighted'))
```
